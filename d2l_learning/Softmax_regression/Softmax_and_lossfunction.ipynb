***Softmax***


```python
import numpy as np
import torch 
import d2l_tools as d2l
```

```python
def softmax(x):
    return np.array([np.exp(x[i]) / sum(np.exp(x)) for i in range(len(x))])
```

```python
x = [1, 3, 5]
softmax(x),sum(softmax(x))
```

***损失函数***

*分析方法：画出Loss_function和gradient图像*

*L2 Loss*


```python
def L2_loss(y_hat, y):
    return (y_hat - y)**2 / 2
```

```python
y_true = 0
y_hat = torch.arange(-10,10,0.01,requires_grad=True)

Loss = L2_loss

Loss(y_hat, y_true).sum().backward()

d2l.plot(y_hat.detach().numpy(), [Loss(y_hat, y_true).detach().numpy(), y_hat.grad],'x','y',legend=['Loss_function','Gradient'])

# 从此图可以看出y_hat远离y_true时，梯度下降速度很快；当两者接近时，下降速度变慢
```

*L1 Loss*


```python
def L1_loss(y_hat, y):
    return abs(y_hat - y)
```

```python
y_true = 0
y_hat = torch.arange(-10,10,0.01,requires_grad=True)

Loss = L1_loss

Loss(y_hat, y_true).sum().backward()

d2l.plot(y_hat.detach().numpy(), [Loss(y_hat, y_true).detach().numpy(), y_hat.grad],'x','y',legend=['Loss_function','Gradient'])

# 可以看出，当y_hat远离y_true时，梯度会均匀下降；当两者接近时，会发生震荡
```

Huber's Robust Loss

其中一种情况


```python
def Huber_Loss(y_hat, y):
    return torch.where(abs(y_hat - y) > 1, abs(y_hat- y) - 1/2, (y_hat - y)**2 / 2)
```

```python
y_true = 0
y_hat = torch.arange(-10,10,0.01,requires_grad=True)

Loss = Huber_Loss

Loss(y_hat, y_true).sum().backward()

d2l.plot(y_hat.detach().numpy(), [Loss(y_hat, y_true).detach().numpy(), y_hat.grad],'x','y',legend=['Loss_function','Gradient'])

# 可看出，此种方法兼备了L1_Loss和L2_Loss的优点 
```
