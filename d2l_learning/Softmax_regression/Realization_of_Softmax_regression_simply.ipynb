```python
import torch
from torch import nn
import d2l_tools as d2l
```

```python
batch_size = 256
train_iter, test_iter = d2l.load_data_fashion_mnist(batch_size)
```

初始化模型参数


```python
# Pytorch 不会隐式地调整输入形状
net = nn.Sequential(nn.Flatten(),nn.Linear(784,10))
def init_weights(m):
    if type(m) == nn.Linear:
        # weight 默认为 0
        nn.init.normal_(m.weight, std=0.01)

net.apply(init_weights)
```

```python
"""
reduction共有三种选项'elementwise_mean'，'sum'和'none'。
'elementwise_mean'为默认情况，表明对N个样本的loss进行求平均之后返回;(reduce=True，size_average=True)
'sum'指对n个样本的loss求和;(reduce=True，size_average=False)
'none'表示直接返回n分样本的loss (reduce=False)

"""
# 调用交叉熵损失函数的时候默认使用softmax激活函数
loss = nn.CrossEntropyLoss(reduction="none")
```

```python
# 优化算法 
trainer = torch.optim.SGD(net.parameters(), lr=0.1)
```

```python
num_epochs = 10
d2l.train_ch3(net, train_iter, test_iter, loss, num_epochs, trainer)
```

***练习***


```python
net.apply(init_weights)
trainer = torch.optim.SGD(net.parameters(), lr=0.05)
num_epochs = 50
d2l.train_ch3(net, train_iter, test_iter, loss, num_epochs, trainer)
```
