```python
import torch
import d2l_tools as d2l
```

```python
# sigmoid 函数
def sigmoid(x):
    return 1 / (1 + torch.exp(-x))

x = torch.arange(-10, 10, 0.01,requires_grad=True)
y = sigmoid(x)
y.sum().backward()

d2l.plot(x.detach(), [y.detach(), x.grad.detach()],'x','y', legend=['sigmoid(x)', 'sigmoid`(x)'])
```

```python
# tanh 函数
def tanh(x):
    return (1 - torch.exp(-2 * x)) / (1 + torch.exp(-2 * x))

x = torch.arange(-10, 10, 0.01,requires_grad=True)
y = tanh(x)
y.sum().backward()
print(x.grad)

d2l.plot(x.detach(), [y.detach(), x.grad.detach()],'x','y', legend=['tanh(x)', 'tanh`(x)'])
```

```python
# RELU 函数
def RELU(x):
    return torch.where(x < 0, 0, x)

x = torch.arange(-10, 10, 0.01,requires_grad=True)
y = RELU(x)
y.sum().backward()

d2l.plot(x.detach(), [y.detach(), x.grad.detach()],'x','y', legend=['RELU(x)', 'RELU`(x)'])
```
